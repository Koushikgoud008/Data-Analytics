# SeaBorn - Statistical Data Visualization Library

### What is seaborn?

SeaBorn is a python library that builds on top of matplotlib that makes the statistical data more easy and Beautiful.

### Why we need it?

1. For creating plots like violin plot, box plot and pair plots.

2. Automatically handles data from pandas dataframes.

### Where we use it?

1. Data Analysis and reporting.

2. Exploring relationships between variables.

3. Machine Learning(for visualizing data sets and results)


## 2. Importing Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = sns.load_dataset("titanic")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.shape
```

### 4. Histograms & Countplots

### Histogram

```python
sns.histplot(df['age'].dropna(), bins=10, kde=True)
plt.title("Histogram of Age")
plt.show()
```

 
### Countplot

```python
sns.countplot(x='class', data=df)
plt.title("Count of Passengers by Class")
plt.show()
```

### 5. Scatter Plots

```python
sns.scatterplot(x='age', y='fare', data=df, hue='sex')
plt.title("Age vs Fare")
plt.show()
 
```

### 6. Boxplot & Violin Plot


### Boxplot

 

```python
sns.boxplot(x='class', y='age', data=df)
plt.title("Boxplot of Age by Class")
plt.show()
```

 
### Violin plot

```python
sns.violinplot(x='class', y='age', hue='sex', data=df, split=True)
plt.title("Violin Plot of Age by Class and Sex")
plt.show()
```

### 7. Pairplot


```python
iris = sns.load_dataset('iris')
sns.pairplot(iris, hue='species')
plt.show()
```

### 8.Heatmap

### Correlation heatmap


```python
corr = iris.select_dtypes(include=['float64', 'int64']).corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap of Iris Dataset")
plt.show()
```

### 9. Customization & Themes

```python
sns.set_theme(style="whitegrid")  # options: darkgrid, white, ticks, etc.
sns.histplot(df['age'].dropna(), bins=10, color='orange')
plt.title("Customized Histogram with Seaborn")
plt.show()
```

```python
sns.barplot(x='class', y='fare', data=df, hue='sex', ci='sd')
plt.title("Average Fare by Class and Sex")
plt.show()
```

```python
sns.pointplot(x='class', y='fare', data=df, hue='sex', markers=['o', 'x'])
plt.title("Point Plot of Fare by Class and Sex")
plt.show()
```

```python
sns.kdeplot(df['fare'].dropna(), fill=True, color='teal')
plt.title("KDE Plot of Fare Distribution")
plt.show()
```

## Hands-On Exercise

1. Compare Passenger Survival by Class and Gender

```python
df.head()
```

```python
sns.set(style="whitegrid")
sns.catplot(data=df,x="class",hue="sex",col="survived",kind="count",height=4,aspect=1)
plt.suptitle("Passenger Survival by Class and Gender", y=1.05)
plt.show()
```

2. Distribution of Passenger Ages

```python
sns.histplot(data=df, x="age", kde=True, bins=30, color="skyblue")
plt.title("Distribution of Passenger Ages")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

3. Fare Distribution by Class

```python
sns.histplot(data=df, x="fare", hue="class", multiple="stack", bins=30)
plt.title("Stacked Fare Distribution by Class")
plt.xlabel("Fare")
plt.ylabel("Count")
plt.show()
```

4. Relationship Between Age and Fare

```python

sns.scatterplot(data=df, x="age", y="fare", hue="class", style="sex", palette="deep")
plt.title("Relationship Between Age and Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

5. Correlation Heatmap

```python
numeric_df = df.select_dtypes(include=["float64", "int64"])
corr_matrix = numeric_df.corr()
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap of Titanic Numeric Features")
plt.show()
```
